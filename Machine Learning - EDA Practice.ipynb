```python
#label aencoding

#import Labelencoder class
from sklearn.preprocessing import LabelEncoder

#create an instance of LabelEncoder
le=LabelEncoder()


#apply label encoding.
df_cat['fuel-type']=le.fit_transform(df_cat["fuel-type"])
```

```python
le.classes_
```

```python
#applying label encoding on all col using loop

#import Labelencoder class
from sklearn.preprocessing import LabelEncoder

for i in df_cat:
     #create an instance of LabelEncoder
    le=LabelEncoder()
    #apply laebel encoding.
    df_cat[i]=le.fit_transform(df_cat[i])
```
